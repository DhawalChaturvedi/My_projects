# Importing necessar libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , f1_score , precision_score , recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

# Importing the data

```python
data = pd.read_csv('creditcard.csv')
```

```python
normal = data[data['Class'] == 0]
fake = data[data['Class'] == 1]
```

```python
normal.describe()
```

```python
fake.describe()
```

```python
normal_sample = normal.sample(n=492)
```

```python
new_data = pd.concat([normal_sample , fake] , axis=0)
```

```python
new_data.head()
```

```python
new_data.tail(6)
```

```python
new_data['Class'].value_counts()               # Now the data is in equal proportion
```

```python
x = new_data.drop(columns='Class' , axis=1)
y = new_data['Class']
```

# Fitting our model

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)
```

```python
classifier = {
    'Logistic regression' : LogisticRegression(),
    'Decision tree classifier' : DecisionTreeClassifier()
}

for n , clf in classifier.items():
    print(f"\n========{n}========")
    clf.fit(x_train , y_train)
    y_pred = clf.predict(x_test)
    print("Accuracy score is :" , accuracy_score(y_test , y_pred))
    print("Precision score is :" , precision_score(y_test , y_pred))    
    print("Recall score is :" , recall_score(y_test , y_pred))    
    print("F1 score is :" , f1_score(y_test , y_pred))    
```

```python
# It's evident that Logistic regression is more precise and gives way more accurate results
```
